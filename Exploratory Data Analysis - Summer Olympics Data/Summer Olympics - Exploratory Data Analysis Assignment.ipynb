<h2>5. Distribution of Participants in Summer Olympics</h2>

<h4>5.1 Distribution of participants in Summer Olympics (1896 - 2012)</h4>

So, let's visualize the distribution of overall athletes participated every year in Summer Olympics (1896 - 2012).

```python
Year_Gender_list = df[["Year", "Gender"]]
Year_paricipant_list = Year_Gender_list.groupby(["Year"]).count()
Year_paricipant_list.plot(kind = "bar", grid = True, figsize=(10,6))
plt.xlabel('Years') # add to x-label to the plot
plt.ylabel('Number of Paticipants') # add y-label to the plot
plt.title('Distribution of participants in Summer Olympics (1896 - 2012)') # add title to the plot
plt.legend(["Participants"])
plt.show()
```

From the visualization, we can able to understand that participants are lesser in the initial seasons of summer olympics and post 1960's the participation of athletes has gradually increased.

Now let's seggregate the Men and Women participants in each season of Summer Olympics (1896 - 2012)

```python
participants_in_each_olympics = df.groupby(['Gender'])['Year'].value_counts()

Men_participants = participants_in_each_olympics.Men.to_frame()
Women_participants = participants_in_each_olympics.Women.to_frame()

Men_participants.rename(columns={'Year': 'Years', "Year":"Participants"}, inplace=True)
Women_participants.rename(columns={'Year': 'Years', "Year":"Participants"}, inplace=True)

Men_participants_sorted = Men_participants.sort_index()
Women_participants_sorted = Women_participants.sort_index()
```

From the visualization we can able to understand that in the recent years the men participants count has reached a saturation level and in 1920 there was a sudden spike in participants count in Summer Olympics.

<h4>7.3 [Bar Chart] Gold vs Silver vs Bronze Medals Split-up of Top Athletes in Summer Olympics (1896 - 2012)</h4>

```python
#Splitting the Medal types and won by the Athletes
Athlete_Medal_Counts = df[["Athlete","Medal"]]
Athlete_Gold_count = Athlete_Medal_Counts[Athlete_Medal_Counts.Medal == 'Gold'].groupby(["Athlete"]).count()
Athlete_Silver_count = Athlete_Medal_Counts[Athlete_Medal_Counts.Medal == 'Silver'].groupby(["Athlete"]).count()
Athlete_Bronze_count = Athlete_Medal_Counts[Athlete_Medal_Counts.Medal == 'Bronze'].groupby(["Athlete"]).count()

#Sorting the Athletes based on their total medal counts
Athlete_Medal_merge = pd.merge(Athlete_Gold_count, Athlete_Silver_count, on='Athlete',how='left')
Athlete_Medal_merge = pd.merge(Athlete_Medal_merge, Athlete_Bronze_count, on='Athlete',how='left')
Athlete_Medal_merge.rename(columns={'Medal_x': 'Gold', "Medal_y":"Silver","Medal" : "Bronze"}, inplace=True)
Athlete_Medal_merge.replace(np.nan,0,inplace=True)
Athlete_Medal_merge = Athlete_Medal_merge.astype("int")
Athlete_Medal_merge["Total"] = Athlete_Medal_merge["Gold"] + Athlete_Medal_merge["Silver"] + Athlete_Medal_merge["Bronze"]
Athlete_Medal_merge_sorted = Athlete_Medal_merge.sort_values("Total", ascending = False).head(20)
Athlete_Medal_merge_sorted = Athlete_Medal_merge_sorted[["Gold","Silver","Bronze"]]

#Visualizing the results
Athlete_Medal_merge_sorted.plot(kind="bar",figsize=(10, 6), grid=True)
plt.xlabel("Athlete Name")
plt.ylabel("Number of Medals")
plt.title("Gold vs Silver vs Bronze Medals Split-up of Top Athletes in Summer Olympics (1896 - 2012)")
plt.show()
```
